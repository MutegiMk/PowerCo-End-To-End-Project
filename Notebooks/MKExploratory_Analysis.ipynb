## EDA *Visualization*

```python
# importing packages for Viz
import matplotlib.pyplot as plt
import seaborn as sns
# Setting plot style
sns.set(color_codes=True)
```

> Plotting stacked bars to particully show the composition of churned customers using **churn** and **sales channels** columns

```python
def plot_stacked_bars(data, title, size=(18, 10), rotation=0, legend="upper right"):
    """
    Generate a stacked bar plot with annotations
    """
    ax = data.plot(
        kind="bar",
        stacked=True,
        figsize=size,
        rot=rotation,
        title=title
    )

    # Add annotations to the bars
    annotate_stacked_bars(ax, text_size=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend)
    # Set labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, padding=0.99, color="white", text_size=13):
    """
    Add value annotations to the stacked bars
    """

    # Iterate over the plotted rectangles/bars
    for rect in ax.patches:
        
        # Calculate the annotation value
        value = str(round(rect.get_height(), 1))
        # Skip annotation if the value is 0
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((rect.get_x() + rect.get_width() / 2) * padding - 0.05, (rect.get_y() + rect.get_height() / 2) * padding),
            color=color,
            size=text_size
        )
```

> Churn overall status

```python
# Select the 'id' and 'churn' columns from the client_data DataFrame
churn = client_data[['id', 'churn']]
# Rename the columns to 'Companies' and 'churn'
churn.columns = ['Companies', 'churn']
# Grouping the churn DataFrame by the 'churn' column and count the occurrences of each category
churn_total = churn.groupby(churn['churn']).count()
# Calculating the percentage of churn for each category by dividing the churn_total by the sum of all values and multiplying by 100
churn_percentage = churn_total / churn_total.sum() * 100
```

```python
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend ="upper right")
```

> From the stacked plot the percentage of churned customers is significantly low

> Sale Channels churn rate

```python
# Selecting the 'id', 'channel_sales', and 'churn' columns from the client DataFrame
channel = client_data[['id', 'channel_sales', 'churn']]

# Grouping the channel DataFrame by the 'channel_sales' and 'churn' columns and count the occurrences of each category in 'id'
# Missing values are filled with 0
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1).fillna(0)

# Calculating the percentage of churn for each channel_sales category by dividing each row by the sum of the row and multiplying by 100
channel_churn = (channel.div(channel.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)
```

```python
plot_stacked_bars(channel_churn, 'Diff Sale channels', rotation=30)
```

> The above plot indicates 5 channels that have churn rate and 3 that have 100% retention

### Visualizing different consumption with respect to churn rate

```python
# Assgning columns to the new dataframe called Consumption
consumption = client_data[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]
```

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Ploting variable distribution in a stacked histogram of churned or retained company
    """
    # Creating a temporary dataframe with the data to be plotted
    retained_data = dataframe[dataframe["churn"] == 0][column]
    churned_data = dataframe[dataframe["churn"] == 1][column]
    temp = pd.DataFrame({"Retention": retained_data, "Churn": churned_data})
    
    # Ploting the stacked histogram
    temp[["Retention", "Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    
    # Setting the x-axis label
    ax.set_xlabel(column)
    
    # Settting the x-axis tick labels to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
# Creating a figure and subplots with a grid of 4 rows and 1 column
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

# Plotting the distribution of 'cons_12m' variable
plot_distribution(consumption, 'cons_12m', axs[0])

# Plotting the distribution of 'cons_gas_12m' variable for entries where 'has_gas' is true
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])

# Plotting the distribution of 'cons_last_month' variable
plot_distribution(consumption, 'cons_last_month', axs[2])

# Plotting the distribution of 'imp_cons' variable
plot_distribution(consumption, 'imp_cons', axs[3])
```

> The consumption data exhibits a significant positive skewness, indicating a pronounced concentration of values towards the lower end of the distribution and a long tail towards higher values. This skewness suggests the presence of outliers, both on the higher and lower ends of the distribution. To gain a better understanding of these outliers, a boxplot can be employed. A boxplot provides a standardized visualization that summarizes the distribution using key summary statistics, including the minimum and maximum values, the median, and the first and third quartiles. By examining the boxplot it is easy to identify and investigate potential outliers in greater detail.

## Hypothesis on Price Sensitivity

> Price sensitivity refers to how responsive customers are to changes in prices. In the context of this analysis, it can be defined as the impact of price changes on the likelihood of churn. It measures how sensitive customers are to price variations and whether those variations influence their decision to churn or stay with the utility company.

>> To evaluate price sensitivity, I will leverage the consumption data available for each company in the year 2015. I will create new features that capture different aspects of price sensitivity based on average consumption over the year, as well as shorter-term trends in the last 6 months and last 3 months. By examining these time frames, I can assess how customers' consumption patterns relate to price changes and identify potential correlations between price sensitivity and churn behavior.

```python
# Converting date columns to date type
date_columns = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']
for col in date_columns:
    client_data[col] = pd.to_datetime(client_data[col])
price_data['price_date'] = pd.to_datetime(price_data['price_date'])
```

```python
# Create mean average data
mean_year = price_data.groupby(['id']).mean().reset_index()
mean_6m = price_data[price_data['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index()
mean_3m = price_data[price_data['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index()

# Comnbine into single dataframe
mean_year = mean_year.rename(
    index=str, 
    columns={
        "price_off_peak_var": "mean_year_price_p1_var",
        "price_peak_var": "mean_year_price_p2_var",
        "price_mid_peak_var": "mean_year_price_p3_var",
        "price_off_peak_fix": "mean_year_price_p1_fix",
        "price_peak_fix": "mean_year_price_p2_fix",
        "price_mid_peak_fix": "mean_year_price_p3_fix"
    }
)

mean_year["mean_year_price_p1"] = mean_year["mean_year_price_p1_var"] + mean_year["mean_year_price_p1_fix"]
mean_year["mean_year_price_p2"] = mean_year["mean_year_price_p2_var"] + mean_year["mean_year_price_p2_fix"]
mean_year["mean_year_price_p3"] = mean_year["mean_year_price_p3_var"] + mean_year["mean_year_price_p3_fix"]

mean_6m = mean_6m.rename(
    index=str, 
    columns={
        "price_off_peak_var": "mean_6m_price_p1_var",
        "price_peak_var": "mean_6m_price_p2_var",
        "price_mid_peak_var": "mean_6m_price_p3_var",
        "price_off_peak_fix": "mean_6m_price_p1_fix",
        "price_peak_fix": "mean_6m_price_p2_fix",
        "price_mid_peak_fix": "mean_6m_price_p3_fix"
    }
)
mean_6m["mean_6m_price_p1"] = mean_6m["mean_6m_price_p1_var"] + mean_6m["mean_6m_price_p1_fix"]
mean_6m["mean_6m_price_p2"] = mean_6m["mean_6m_price_p2_var"] + mean_6m["mean_6m_price_p2_fix"]
mean_6m["mean_6m_price_p3"] = mean_6m["mean_6m_price_p3_var"] + mean_6m["mean_6m_price_p3_fix"]

mean_3m = mean_3m.rename(
    index=str, 
    columns={
        "price_off_peak_var": "mean_3m_price_p1_var",
        "price_peak_var": "mean_3m_price_p2_var",
        "price_mid_peak_var": "mean_3m_price_p3_var",
        "price_off_peak_fix": "mean_3m_price_p1_fix",
        "price_peak_fix": "mean_3m_price_p2_fix",
        "price_mid_peak_fix": "mean_3m_price_p3_fix"
    }
)
mean_3m["mean_3m_price_p1"] = mean_3m["mean_3m_price_p1_var"] + mean_3m["mean_3m_price_p1_fix"]
mean_3m["mean_3m_price_p2"] = mean_3m["mean_3m_price_p2_var"] + mean_3m["mean_3m_price_p2_fix"]
mean_3m["mean_3m_price_p3"] = mean_3m["mean_3m_price_p3_var"] + mean_3m["mean_3m_price_p3_fix"]

# Merge into 1 dataframe
price_features = pd.merge(mean_year, mean_6m, on='id')
price_features = pd.merge(price_features, mean_3m, on='id')
```

```python
price_features.head()
```

### Correlation between Price sensitivity and customer churn

```python
#merging dataset price features data set with client data
price_sensitivity = pd.merge(price_features, client_data[['id', 'churn']], on='id')
price_sensitivity.head()
```

```python
corr = price_sensitivity.corr()
# Plotting correlation
plt.figure(figsize=(20,18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot = True, annot_kws={'size':10})
# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

> The correlation analysis reveals that the price sensitivity variables exhibit a relatively strong correlation among themselves, indicating a higher level of association. However, when considering the correlation with the churn variable, it is observed to be quite low. This suggests a weak linear relationship between price sensitivity and churn. Thus, relying solely on price sensitivity as a primary predictor for churn may not be sufficient. It is recommended to explore alternative feature engineering techniques to enhance the predictive power of price sensitivity in predicting churn.
